```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
from PIL import Image
```

```python

img = cv2.imread('PHONE.png')
median = cv2.medianBlur(img,5)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Mediana')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

royal_1 = np.array(Image.open('IMG_1335.jpg'))
royal_2 = np.array(Image.open('IMG_1336.jpg'))
royal_3 = np.array(Image.open('IMG_1337.jpg'))
royal_4 = np.array(Image.open('IMG_1338.jpg'))
royal_avg = ( royal_1.astype(np.float64)
            + royal_2.astype(np.float64)
            + royal_3.astype(np.float64)
            + royal_4.astype(np.float64)) / 4
royal_avg = royal_avg.astype(np.uint8)
Image.fromarray(royal_avg).save('royal_avg.jpg')

plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(royal_1)
plt.subplot(2, 2, 2)
plt.imshow(royal_2)
plt.subplot(2, 2, 3)
plt.imshow(royal_3)
plt.subplot(2, 2, 4)
plt.imshow(royal_avg)
```
